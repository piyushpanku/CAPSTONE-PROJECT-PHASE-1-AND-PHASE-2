```python
import csv
import requests
from bs4 import BeautifulSoup

# Function to get movie details from the provided IMDb URL
def get_movie_details(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    movies = soup.find_all("div", class_="lister-item-content")

    movie_data = []
    for movie in movies:
        # Movie Details
        movie_name = movie.h3.a.text
        duration = movie.find("span", class_="runtime").text.strip(" min")
        year = movie.find("span", class_="lister-item-year").text.strip("()")
        rating = movie.find("div", class_="inline-block ratings-imdb-rating").strong.text
        metascore = movie.find("span", class_="metascore").text.strip() if movie.find("span", class_="metascore") else ""

        # Directors
        directors = movie.find_all("p")[2].find("a")
        director_names = [director.text for director in directors]

        # Stars
        stars = movie.find_all("p")[2].find_all("a")[1:]
        star_names = [star.text for star in stars]

        # Other Details
        votes_element = movie.find("span", attrs={"name": "nv"})
        votes = votes_element["data-value"] if votes_element else ""

        genre_element = movie.find("span", class_="genre")
        genre = genre_element.text.strip() if genre_element else ""

        gross_element = movie.find("p", class_="sort-num_votes-visible")
        gross = gross_element.find_all("span")[-1].text.strip("M") if gross_element else ""

        popularity_element = movie.find("div", class_="inline-block ratings-imdb-rating")
        popularity = popularity_element["data-tconst"] if popularity_element and "data-tconst" in popularity_element else ""

        certification_element = movie.find("span", class_="certificate")
        certification = certification_element.text.strip() if certification_element else ""

        # Append data to movie_data list
        movie_data.append([movie_name, director_names, duration, year, rating, metascore, star_names, votes, genre, gross, popularity, certification])

    return movie_data

# Function to write data to CSV file
def write_to_csv(data, headers, filename):
    with open(filename, mode="w", encoding="utf-8", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(data)

# IMDb URL for action genre movies with a minimum of 25,000 votes sorted by user rating
url = "https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1"
movie_data = get_movie_details(url)

# Headers for CSV files
headers1 = ["Sno", "Movie Name", "Director1", "Director2", "Duration", "Year", "Ratings", "Metascore"]
headers2 = ["Movie Name", "Star1", "Star2", "Star3", "Star4", "Votes", "Genre1", "Genre2", "Genre3", "Gross Collection", "Popularity", "Certification"]

# Prepare data for CSV file 1
csv_data1 = []
for idx, movie in enumerate(movie_data, 1):
    sno = idx
    movie_name = movie[0]
    director1 = movie[1][0] if len(movie[1]) > 0 else ""
    director2 = movie[1][1] if len(movie[1]) > 1 else ""
    duration = movie[2]
    year = movie[3]
    ratings = movie[4]
    metascore = movie[5]
    csv_data1.append([sno, movie_name, director1, director2, duration, year, ratings, metascore])

# Prepare data for CSV file 2
csv_data2 = []
for movie in movie_data:
    movie_name = movie[0]
    star1 = movie[6][0] if len(movie[6]) > 0 else ""
    star2 = movie[6][1] if len(movie[6]) > 1 else ""
    star3 = movie[6][2] if len(movie[6]) > 2 else ""
    star4 = movie[6][3] if len(movie[6]) > 3 else ""
    votes = movie[7]
    genres = movie[8].split(", ")
    genre1 = genres[0] if len(genres) > 0 else ""
    genre2 = genres[1] if len(genres) > 1 else ""
    genre3 = genres[2] if len(genres) > 2 else ""
    gross_collection = movie[9]
    popularity = movie[10]
    certification = movie[11]
    csv_data2.append([movie_name, star1, star2, star3, star4, votes, genre1, genre2, genre3, gross_collection, popularity, certification])

# Write data to CSV files
write_to_csv(csv_data1, headers1, "movie_details.csv")
write_to_csv(csv_data2, headers2, "movie_other_details.csv")
```

```python
import pandas as pd
import sqlite3

# Load the CSV data into Pandas data frames
dataframe1 = pd.read_csv("movie_details.csv")
dataframe2 = pd.read_csv("movie_other_details.csv")

# Create an SQLite database connection
conn = sqlite3.connect("imdb_database.db")

# Insert data into the tables
dataframe1.to_sql("Table1", conn, if_exists="replace", index=False)
dataframe2.to_sql("Table2", conn, if_exists="replace", index=False)

# Function to perform queries using Pandas SQL
def pandas_sql_queries():
    # Task 1
    query1 = "SELECT * FROM Table1 WHERE Director1 IN ('Christopher', 'Matt Reeves')"
    result1 = pd.read_sql_query(query1, conn)

    # Task 2
    query2 = "SELECT * FROM Table1 WHERE Duration BETWEEN 140 AND 190"
    result2 = pd.read_sql_query(query2, conn)

    # Task 3
    query3 = "SELECT * FROM Table1 WHERE Ratings > 7 ORDER BY Ratings ASC"
    result3 = pd.read_sql_query(query3, conn)

    # Task 4
    query4 = "SELECT * FROM Table1 ORDER BY [Movie Name] DESC"
    result4 = pd.read_sql_query(query4, conn)

    # Task 5
    query5 = "SELECT * FROM Table1 WHERE [Movie Name] LIKE 'P%' AND Ratings > 7"
    result5 = pd.read_sql_query(query5, conn)

    # Task 6
    query6 = "SELECT * FROM Table2 WHERE star1 LIKE '%Arnold Schwarzenegger%' ORDER BY [Movie Name] ASC"
    result6 = pd.read_sql_query(query6, conn)

    # Task 7
    query7 = "SELECT * FROM Table2 ORDER BY votes DESC LIMIT 1"
    result7 = pd.read_sql_query(query7, conn)

    # Task 8
    query8 = "SELECT * FROM Table2 ORDER BY [Gross collection] DESC"
    result8 = pd.read_sql_query(query8, conn)

    # Task 9
    query9 = "SELECT * FROM Table2 WHERE star1 LIKE '%Arnold%' AND [Gross collection] IS NOT NULL"
    result9 = pd.read_sql_query(query9, conn)

    # Task 10
    query10 = "SELECT * FROM Table2 WHERE Genre1 LIKE '%comedy%' OR Genre1 LIKE '%action%'"
    result10 = pd.read_sql_query(query10, conn)

    # Task 11
    query11 = "SELECT Table1.*, Table2.* FROM Table1 JOIN Table2 ON Table1.[Movie Name] = Table2.[Movie Name]"
    result11 = pd.read_sql_query(query11, conn)

    # Task 12
    query12 = "SELECT Table1.[Movie Name], Table1.Director1, Table1.ratings, Table2.[Gross collection] " \
              "FROM Table1 JOIN Table2 ON Table1.[Movie Name] = Table2.[Movie Name] WHERE Table2.Genre1 LIKE '%action%'"
    result12 = pd.read_sql_query(query12, conn)

    # Task 13
    query13 = "SELECT Table1.*, Table2.* FROM Table1 JOIN Table2 ON Table1.[Movie Name] = Table2.[Movie Name] " \
              "ORDER BY CAST(REPLACE(Table2.[Gross collection], 'M', '') AS FLOAT) DESC LIMIT 1"
    result13 = pd.read_sql_query(query13, conn)

    # Task 14
    query14 = "SELECT Table1.*, Table2.* FROM Table1 JOIN Table2 ON Table1.[Movie Name] = Table2.[Movie Name] " \
              "ORDER BY CAST(Table1.Ratings AS FLOAT) DESC LIMIT 1"
    result14 = pd.read_sql_query(query14, conn)

    # Task 15
    query15 = "SELECT Table1.*, Table2.* FROM Table1 JOIN Table2 ON Table1.[Movie Name] = Table2.[Movie Name] " \
              "ORDER BY CAST(REPLACE(Table2.[Gross collection], 'M', '') AS FLOAT) ASC, " \
              "CAST(Table1.Ratings AS FLOAT) ASC LIMIT 1"
    result15 = pd.read_sql_query(query15, conn)

    # Close the connection
    conn.close()

    return result1, result2, result3, result4, result5, result6, result7, result8, result9, result10, \
           result11, result12, result13, result14, result15

# Perform queries using Pandas SQL
result1, result2, result3, result4, result5, result6, result7, result8, result9, result10, \
    result11, result12, result13, result14, result15 = pandas_sql_queries()
```
